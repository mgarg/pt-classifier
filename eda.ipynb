```python
from datasets import load_dataset
dataset = load_dataset(
    "parquet",
    data_files={
        "train": "hf://datasets/McAuley-Lab/Amazon-Reviews-2023/raw_meta_All_Beauty/full-00000-of-00001.parquet"
    },
    split="train"
)

dataset = dataset.shuffle(seed = 42).select(range(10000))
dataset.to_parquet("data/sampled/sample.raw.10k.parquet")
dataset.to_json("data/sampled/sample.json.sample_10k.jsonl", lines = True)
```
